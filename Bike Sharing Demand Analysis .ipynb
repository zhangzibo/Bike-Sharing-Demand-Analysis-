```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings("ignore")
pd.options.display.max_columns =999
```

```python
df = df.rename(columns={'weathersit':'weather',
                       'yr':'year',
                       'mnth':'month',
                       'hr':'hour',
                       'hum':'humidity',
                       'cnt':'count'})
df.head()
```

```python
df = df.drop(columns=["instant","dteday","year"], axis=0)
```

```python
cols = ["season", "month","hour","holiday","weekday","workingday","weather"]

for col in cols:
    df[col] = df[col].astype("category")
df.info()
```

```python
fig,ax = plt.subplots(figsize=(20,10))

sns.pointplot(data=df,x="hour",y="count",hue = "weekday",ax=ax)
ax.set(title="count of bike during weekdays")
```

```python
fig,ax = plt.subplots(figsize=(20,10))

sns.pointplot(data=df,x="hour",y="casual",hue = "weekday",ax=ax)
ax.set(title="count of bike during weekdays: unregister")
```
